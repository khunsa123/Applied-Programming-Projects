```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import date2num
import pandas as pd
from datetime import datetime

# Define the tasks and start/end dates
tasks = [
    ("Literature Review & Data Collection", "2024-01-01", "2024-03-31"),
    ("Model Development (First Iteration)", "2024-04-01", "2024-06-30"),
    ("Initial Model Testing & Optimization", "2024-07-01", "2024-09-30"),
    ("Real-World Data Collection & Evaluation", "2024-10-01", "2024-12-31"),
    ("Model Refinement Based on User Feedback", "2025-01-01", "2025-03-31"),
    ("Present Interim Findings & User Testing", "2025-04-01", "2025-06-30"),
    ("Usability Improvements & Large-Scale Testing", "2025-07-01", "2025-09-30"),
    ("Final Model Refinement & Publication", "2025-10-01", "2025-12-31"),
    ("Thesis Writing & Submission", "2026-01-01", "2026-03-31"),
    ("Final Adjustments & Thesis Presentation", "2026-04-01", "2026-06-30")
]

# Create a dataframe
df = pd.DataFrame(tasks, columns=["Task", "Start", "End"])

# Convert dates
df["Start"] = pd.to_datetime(df["Start"])
df["End"] = pd.to_datetime(df["End"])

# Initialize the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting each task as a horizontal bar
for i, (task, start, end) in enumerate(zip(df["Task"], df["Start"], df["End"])):
    ax.barh(task, (end - start).days, left=date2num(start), height=0.4, color="skyblue")

# Format the x-axis to show dates properly
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b-%Y"))
plt.xticks(rotation=45)

# Labels and title
ax.set_xlabel("Time")
ax.set_ylabel("Tasks")
ax.set_title("Gantt Chart for PhD Research Timeline")

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates

# Define tasks and timelines
tasks = ['Literature Review', 'Data Collection & Preprocessing', 'Model Development', 'Model Optimization', 'Testing & Evaluation', 'User Feedback & Refinement', 'Thesis Writing']
start_dates = ['2024-01-01', '2024-04-01', '2024-07-01', '2024-10-01', '2025-01-01', '2025-06-01', '2025-10-01']
end_dates = ['2024-03-31', '2024-06-30', '2024-09-30', '2024-12-31', '2025-05-31', '2025-09-30', '2025-12-31']

# Convert the dates to datetime format
start_dates = pd.to_datetime(start_dates)
end_dates = pd.to_datetime(end_dates)

# Calculate the duration for each task
durations = end_dates - start_dates

# Create a Gantt chart
fig, ax = plt.subplots(figsize=(10, 6))

# Iterate over the tasks and plot each one
for i, task in enumerate(tasks):
    ax.barh(task, durations[i].days, left=start_dates[i], height=0.4, align='center')

# Format the x-axis
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

# Add labels and title
plt.xlabel('Timeline')
plt.ylabel('Tasks')
plt.title('PhD Project Timeline')

plt.grid(True)
plt.show()
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
from matplotlib.dates import date2num
from datetime import datetime

# Define timeline tasks and durations
tasks = [
    {"Task": "Literature review and baseline data collection", "Start": "2025-05-01", "End": "2025-12-31"},
    {"Task": "Initial model development for Wi-Fi-based HAR", "Start": "2025-06-01", "End": "2025-12-31"},
    {"Task": "Address noise issues and incorporate attention mechanisms", "Start": "2026-01-01", "End": "2026-06-30"},
    {"Task": "Begin testing multi-modal approaches", "Start": "2026-07-01", "End": "2026-12-31"},
    {"Task": "Validation in real-world applications", "Start": "2027-01-01", "End": "2027-06-30"},
    {"Task": "Refinement of multi-modal techniques", "Start": "2027-07-01", "End": "2027-12-31"},
    {"Task": "Thesis writing and journal publications", "Start": "2028-01-01", "End": "2028-06-30"}
]

# Convert dates to matplotlib format
for task in tasks:
    task["Start"] = date2num(datetime.strptime(task["Start"], "%Y-%m-%d"))
    task["End"] = date2num(datetime.strptime(task["End"], "%Y-%m-%d"))

# Plot Gantt chart
fig, ax = plt.subplots(figsize=(12, 6))

for i, task in enumerate(tasks):
    ax.barh(
        y=i,
        width=task["End"] - task["Start"],
        left=task["Start"],
        color="skyblue",
        edgecolor="black"
    )

# Set labels and grid
ax.set_yticks(range(len(tasks)))
ax.set_yticklabels([task["Task"] for task in tasks])
ax.set_xlabel("Timeline")
ax.xaxis_date()
ax.grid(True, which="both", linestyle="--", alpha=0.5)

# Add title and layout
plt.title("PhD Research Timeline: Human Activity Recognition")
plt.tight_layout()
plt.show()
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
from matplotlib.dates import date2num
from matplotlib.patches import Patch
from datetime import datetime
import numpy as np

# Define timeline tasks and durations
tasks = [
    {"Task": "Literature review and baseline data collection", "Start": "2025-05-01", "End": "2025-12-31"},
    {"Task": "Initial model development for Wi-Fi-based HAR", "Start": "2025-06-01", "End": "2025-12-31"},
    {"Task": "Address noise issues and incorporate attention mechanisms", "Start": "2026-01-01", "End": "2026-06-30"},
    {"Task": "Begin testing multi-modal approaches", "Start": "2026-07-01", "End": "2026-12-31"},
    {"Task": "Validation in real-world applications", "Start": "2027-01-01", "End": "2027-06-30"},
    {"Task": "Refinement of multi-modal techniques", "Start": "2027-07-01", "End": "2027-12-31"},
    {"Task": "Thesis writing and journal publications", "Start": "2028-01-01", "End": "2028-06-30"}
]

# Convert dates to matplotlib format
for task in tasks:
    task["Start"] = date2num(datetime.strptime(task["Start"], "%Y-%m-%d"))
    task["End"] = date2num(datetime.strptime(task["End"], "%Y-%m-%d"))

# Define custom colors
colors = plt.cm.viridis(np.linspace(0.2, 0.8, len(tasks)))

# Plot Gantt chart
fig, ax = plt.subplots(figsize=(14, 7))
for i, task in enumerate(tasks):
    ax.barh(
        y=i,
        width=task["End"] - task["Start"],
        left=task["Start"],
        color=colors[i],
        edgecolor="black",
        alpha=0.9
    )
    # Add task labels inside the bars
    ax.text(
        x=task["Start"] + (task["End"] - task["Start"]) / 2,
        y=i,
        s=task["Task"],
        va="center",
        ha="center",
        color="white",
        fontsize=10,
        fontweight="bold"
    )

# Set labels and grid
ax.set_yticks(range(len(tasks)))
ax.set_yticklabels([f"Task {i+1}" for i in range(len(tasks))], fontsize=10)
ax.set_xlabel("Timeline", fontsize=12)
ax.xaxis_date()
ax.grid(axis="x", linestyle="--", alpha=0.6)

# Add title and legend
plt.title("PhD Research Timeline: Human Activity Recognition", fontsize=14, fontweight="bold")
legend_elements = [Patch(facecolor=colors[i], edgecolor='black', label=tasks[i]["Task"]) for i in range(len(tasks))]
plt.legend(handles=legend_elements, loc='upper left', bbox_to_anchor=(1, 1), title="Tasks", fontsize=10)

# Adjust layout for better appearance
plt.tight_layout()
plt.show()
```
